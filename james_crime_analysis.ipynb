```
import pandas as pd

data = pd.read_csv("http://cluster.earlham.edu/~apc0013/Chicago_crime.csv")
```

```
data.info()
```

```
data.head(10)
```

```
# Locations of crime

groupby_loc = data.groupby(["Location Description"])["ID"].count().sort_values(ascending=False)
groupby_loc[:10].plot(kind="bar", title="10 Locations Where Crimes Occur the Most")
```

```
# Types of crime where location is "Street"
street_crimes = data.loc[data["Location Description"] == "STREET"]
type_street_crimes = street_crimes.groupby("Primary Type")["ID"].count().sort_values(ascending=False)
type_street_crimes[:10].plot(kind="bar", title="10 Most Frequent Street Crimes")
```
