```python
import pandas as pd
import geopandas as gpd
import descartes
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
df = pd.read_csv("http://cluster.earlham.edu/~apc0013/Chicago_crime.csv")
```

```python
pd.set_option('display.max_columns', 500)
df['Date'] = pd.to_datetime(df['Date'])  
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df['Primary Type'].value_counts()
```

```python
# Create Point objects that describe the longitude and latitude of data-points.
geometry = [Point(xy) for xy in zip(df["Longitude"], df["Latitude"])]

# Projection
crs = {'init' :'epsg:4326'}
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
geo_df.head()

# drop rows with abnormal latitude and longtitude
geo_df = geo_df[geo_df.Latitude > 41] 
geo_df = geo_df[geo_df.Longitude > -89]
```

```python
# load the shapefile
shpfile = gpd.read_file('geo_export_96e7b464-746d-4468-9d47-6cf675b6bca6.shp')
shpfile['area_num_1'] = shpfile['area_num_1'].astype(int)
shpfile.head()
```

```python
area_groupby = geo_df.groupby('Community Area')
df_count = area_groupby['ID'].count().to_frame().reset_index()
df_count.columns = ['area_num_1', 'count']
df_count['area_num_1'] = df_count['area_num_1'].astype(int)
```

```python
merged = pd.merge(shpfile, df_count, on='area_num_1', how='inner')
```

```python
# set the range for the choropleth
vmin, vmax = min(merged['count']), max(merged['count'])
# create figure and axes for Matplotlib
fig, ax = plt.subplots(figsize=(15, 12))
# create map
merged.plot('count', ax=ax, cmap= 'Blues', edgecolor='0.9')
ax.axis('off')
# add a title
ax.set_title('Number of crimes by community district', fontdict={'fontsize': '25'})

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
```

```python
years = [2016, 2017, 2018]

counter = 0
for year in years:
    df_year = geo_df[geo_df['Date'].dt.year == year]
    df_year_groupby = df_year.groupby('Community Area')
    df_year_count = df_year_groupby['ID'].count().to_frame().reset_index()
    df_year_count.columns = ['area_num_1', 'count']
    df_year_count['area_num_1'] = df_year_count['area_num_1'].astype(int)
    year_merged = pd.merge(shpfile, df_year_count, on='area_num_1', how='inner')
    
    #vmin, vmax = min(year_merged['count']), max(year_merged['count'])
    fig, ax = plt.subplots(figsize=(12, 10))
    year_merged.plot('count', ax=ax, cmap= 'Blues', edgecolor='0.9')
    ax.axis('off')
    ax.set_title('Number of crimes by community district for {}'.format(year), fontdict={'fontsize': '20'})

    sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=0, vmax=18000))
    #sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm._A = []
    cbar = fig.colorbar(sm)
    
    plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,15))
shpfile.plot(ax = ax, alpha = 0.4, color = "grey", edgecolor='0.9')
geo_df[geo_df['Primary Type'] == "SEX OFFENSE"].plot(ax = ax, markersize = 5, color = 'green', label = "SEX OFFENSE")
plt.legend(prop={"size": 15})
```
