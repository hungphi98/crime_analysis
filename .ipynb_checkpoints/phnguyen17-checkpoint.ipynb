```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("http://cluster.earlham.edu/~apc0013/Chicago_crime.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
labels = ["THEFT", "BATTERY","CRIMINAL DAMAGE","ASSAULT", "DECEPTIVE PRACTICE"]
df = df[df["Primary Type"].isin(labels)]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.shape
```

```python
data = df[df["Date"] > pd.to_datetime("12/31/2017")]
```

```python
data["hour"] = data["Date"].apply(lambda x: x.hour)
```

```python
fig, ax = plt.subplots(figsize=(16,7))
data.groupby("hour").count()["Arrest"].plot(kind="bar", ax = ax)
```

```python
data.groupby("Primary Type").count()["Arrest"].sort_values(ascending = False).head(12).plot(kind="bar")
```

```python
data.groupby("Primary Type").count()["Arrest"].sort_values(ascending = False)
```

```python
fig, ax = plt.subplots(figsize=(16,7))

data.groupby(["hour", "Primary Type"]).count()["Arrest"].unstack('Primary Type').plot(kind = "bar", stacked = True, ax = ax)

```

```python
data.columns
```

```python
data.drop(["ID", "Case Number", "Date", "Block", "IUCR", "Description", "Arrest", "Domestic", \
           "Beat", "Ward", "FBI Code", "X Coordinate", "Y Coordinate", "Year", "Updated On", \
          "Latitude", "Longitude", "Location"],axis = 1)
```

```python
y = data["Primary Type"]
y.shape
```

```python
X =data.drop(["ID", "Case Number", "Date", "Block", "IUCR", "Description", "Arrest", "Domestic", \
           "Beat", "Ward", "FBI Code", "X Coordinate", "Y Coordinate", "Year", "Updated On", \
          "Latitude", "Longitude", "Location", "Primary Type"],axis = 1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation 
from keras import optimizers

output_dim = len(labels)
model = Sequential()
model.add(Dense(8,kernel_initializer='uniform',input_shape = (4,)))
model.add(Dense(32))
model.add(Activation('softmax'))
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='mean_squared_error', optimizer=sgd)
model.fit(X, y,batch_size = 100, epochs = 15, validation_data=(X_test, y_test))
```
